```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def CompareTwo(sample1Idx, sample2Idx):
    sample1Filename="perf_"+str(sample1Idx)+".data"
    sample2Filename="perf_"+str(sample2Idx)+".data"
    sample1Data=pd.read_csv(sample1Filename, sep="\n", header=None)
    sample2Data=pd.read_csv(sample2Filename, sep="\n", header=None)

    x=np.array([100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000,2100,2200,2300,2400,2500,2600,2700,2800,2900,3000])

    length1=sample1Data.shape[0]
    length2=sample1Data.shape[0]
    xlength=min(length1, length2)-1
    x=x[0:xlength]
    plt.plot(x, sample1Data[1:xlength+1] , "*-", label="kernel"+str(sample1Idx))
    plt.plot(x, sample2Data[1:xlength+1] , "d-", label="kernel"+str(sample2Idx))
    plt.legend()
    plt.xlabel("matrix size (m=n=k)")
    plt.ylabel("gflops")

    outputFilename=".\\fig\\perfComp_"+str(sample1Idx)+"_"+str(sample2Idx)+".png"
    plt.savefig(outputFilename)



```

```python
def CompareThree(sample1Idx, sample2Idx, sample3Idx):
    sample1Filename="perf_"+str(sample1Idx)+".data"
    sample2Filename="perf_"+str(sample2Idx)+".data"
    sample3Filename="perf_"+str(sample3Idx)+".data"
    sample1Data=pd.read_csv(sample1Filename, sep="\n", header=None)
    sample2Data=pd.read_csv(sample2Filename, sep="\n", header=None)
    sample3Data=pd.read_csv(sample3Filename, sep="\n", header=None)
    
    x=np.array([100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000,2100,2200,2300,2400,2500,2600,2700,2800,2900,3000])

    length1=sample1Data.shape[0]
    length2=sample1Data.shape[0]
    length3=sample1Data.shape[0]
    
    xlength=min(length1, length2, length3)-1
    x=x[0:xlength]
    plt.plot(x, sample1Data[1:xlength+1] , "*-", label="kernel"+str(sample1Idx))
    plt.plot(x, sample2Data[1:xlength+1] , "d-", label="kernel"+str(sample2Idx))
    plt.plot(x, sample3Data[1:xlength+1] , "s-", label="kernel"+str(sample3Idx))
    plt.legend()
    plt.xlabel("matrix size (m=n=k)")
    plt.ylabel("gflops")

    outputFilename=".\\fig\\perfComp_"+str(sample1Idx)+"_"+str(sample2Idx)+"_"+str(sample3Idx)+".png"
    plt.savefig(outputFilename)

```

```python
CompareThree(12,13,14)
```

```python
CompareTwo(13,15)
```
